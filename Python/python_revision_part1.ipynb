```python
#Part 2 — Lists
#Q3

#Create a list of your subjects.

subjects=["Maths","AI","DBMS","SQL","Aptitude"]
print(subjects,subjects.append("Data Handling"))
print(subjects,subjects.remove("DBMS"))
print(subjects,subjects.sort())
print(subjects,subjects.reverse())
print(len(subjects))
```

```python
#Part 3 — Tuples
#Q5

#Create a tuple of your birth details.
birth = (12, "May", 2004)
print(birth[0])
print(birth[-1])
print(2004 in birth)
```

```python
'''Part 4 — Dictionaries
Q6

Create your student profile:'''

student = {
    "name": "Lixin Josh",
    "cgpa": 8.0,
    "goal": "Germany",
    "work": "Data Engineer"
}
print(student.keys())
print(student.values())
student.update({"skills":"Sql"})
print(student)
student.update({"cgpa": 8.5})
print(student)
student.pop("work")
print(student)
```

```python
#Part 5 — List Comprehensions
#Q7

#Create a list from 1 to 10.

print([x**2 for x in range(1,10)])
print([x for x in range(1,11) if x%2==0])
print([x for x in range(1,11) if x>5])
```
