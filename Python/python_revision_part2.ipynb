```python
#Q1 — Student Eligibility Function

#Create a function:

def check_cgpa(cgpa):
    if cgpa>7.5: return "Eligible for higher studies"
    else: return "Improve academics and Projects"

print(check_cgpa(8.0))
print(check_cgpa(7.0))
```

```python
#Q2 — Marks Average Function

#Create a function that takes a list of marks and returns the average.

def average_marks(marks):
    return sum(marks)/len(marks)
print(average_marks([85, 90, 78, 92, 88]))
print(average_marks([70, 75, 80, 65, 90]))
```

```python
#Q3 — Word Counter

def word_count(sentence):
    words=sentence.split()
    total_words=len(words)
    total_count_wo_space=len(sentence.replace(" ",""))
    total_count=len(sentence)
    return total_count, total_words, total_count_wo_space

total_count, total_words, total_count_wo_space = word_count("I want to become a data engineer")
print(total_count)
print(total_words)
print(total_count_wo_space)
```

```python
#Q6 — Data Filtering

def passed_student(marks):
    passed_students=[]
    for x in range(len(marks)):
        if marks[x]>60:
             passed_students.append(marks[x])
    return passed_students
print(passed_student([45, 67, 89, 32, 91, 76])) 
```

```python
def passed_student(marks):
    passed_students=[x for x in marks if x>60]
    return passed_students
print(passed_student([45, 67, 89, 32, 91, 76]))
```
