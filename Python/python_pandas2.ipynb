```python
'''DAY 5 — Real Data Cleaning + Missing Values

This is very important in Data Engineering.

Because real company data is often:

incomplete
messy
duplicated
inconsistent

Your job is to clean it.'''
```

```python
import pandas as pd

df = pd.read_csv("employees.csv")
print(df)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
age_avg = df["Age"].mean()
df["Age"] = df["Age"].fillna(age_avg)
print(df)
```

```python
avg_salary=df["Salary"].mean()
df["Salary"]=df["Salary"].fillna(avg_salary)
print(df)
```

```python
df.drop_duplicates()
```

```python
df.to_csv("cleaned_employees.csv", index=False)
```
